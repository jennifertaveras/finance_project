# CFPB Consumer Complaints
###cleaning the dataset for analysis

```python
import pandas as pd
df = pd.read_csv("/content/complaints-2026-01-13_16_28.csv")   # path relative to workspace
df.head()
```

```python
df.info()
```

create a manageable raw subset deleteting the columns that doesn't need for analysis

```python
df_subset = df.drop(columns=[
    'Sub-product',
    'Sub-issue',
    'Consumer complaint narrative',
    'Company public response',
    'Tags',
    'Submitted via',
    'Consumer disputed?',
    'Consumer consent provided?'

])
df_subset.head()

```

standard columns

```python
# renaming columns for readability, consistency and data quality
df_subset = df_subset.rename(columns={
    'Date received': 'date_received',
    'Product': 'product',
    'Issue': 'issue',
    'Company': 'company',
    'State': 'state',
    'ZIP code': 'zip_code',
    'Date sent to company': 'date_sent_to_company',
    'Company response to consumer': 'company_response_to_consumer',
    'Timely response?': 'timely_response',
    'Complaint ID': 'complaint_id'
})

df_subset.head()
```

drop duplicates

```python
df_subset = df_subset.drop_duplicates()
df_subset.head()
```

correct date ranges

```python
date_columns = ['date_received', 'date_sent_to_company']

for col in date_columns:
    # Ensure column is in datetime format, coercing errors
    df_subset[col] = pd.to_datetime(df_subset[col], format='%m/%d/%y', errors='coerce')

    print(f"{col} column after specifying format:")
    display(df_subset[col].head())

    # Check for any NaT values that resulted from coercion
    num_invalid_dates = df_subset[col].isna().sum()
    if num_invalid_dates > 0:
        print(f"\nWarning: {num_invalid_dates} dates in '{col}' could not be parsed with the specified format and were converted to NaT.")
    print("\n" + "-"*50 + "\n") # Separator for readability
```

Missing Values

```python
df_subset.isna().sum()
```

```python
df_subset = df_subset.dropna()
```

Product Normalization

```python
df_subset['product'].value_counts()
```

Created a normalize_product function to standardize variations (e.g., grouping different 'credit reporting' entries).

```python
def normalize_product(product):
    product = str(product).lower()
    if 'credit reporting' in product:
        return 'Credit Reporting'
    elif 'credit card' in product:
        return 'Credit Card'
    elif 'payday loan' in product or 'title loan' in product or 'personal loan' in product:
        return 'Payday/Personal/Title Loan'
    elif 'debt collection' in product:
        return 'Debt Collection'
    elif 'checking' in product or 'savings' in product:
        return 'Checking or Savings Account'
    elif 'money transfer' in product or 'virtual currency' in product or 'money service' in product:
        return 'Money Transfer/Virtual Currency'
    elif 'mortgage' in product:
        return 'Mortgage'
    elif 'vehicle loan' in product or 'vehicle lease' in product:
        return 'Vehicle Loan/Lease'
    elif 'student loan' in product:
        return 'Student Loan'
    elif 'prepaid card' in product:
        return 'Prepaid Card'
    elif 'debt or credit management' in product:
        return 'Debt/Credit Management'
    else:
        return product.title() # Default for other products

df_subset = df_subset.copy() # Explicitly create a copy to avoid SettingWithCopyWarning
df_subset['product'] = df_subset['product'].apply(normalize_product)

print("Unique values and counts for 'product' column after normalization:")
print(df_subset['product'].value_counts())
```
